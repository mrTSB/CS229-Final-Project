```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv('data116.csv')

df.head()
```

```python
unique_dates = df['Date'].unique()

print("Unique dates:")
print(unique_dates)
```

```python
# Filter the dataframe for dates up to and including 2022-09-21
chunk1 = df[df['Date'] <= '2022-09-21']

# Display the shape and first few rows of the filtered dataframe
print("Shape of filtered dataframe:", chunk1.shape)
print("\nFirst few rows:")
display(chunk1.head())
```

```python
# Filter the dataframe for dates between 2023-03-30 and 2023-08-03 (inclusive)
chunk2 = df[(df['Date'] >= '2023-03-30') & (df['Date'] <= '2023-08-03')]

# Display the shape and first few rows of the filtered dataframe
print("Shape of filtered dataframe:", chunk2.shape)
print("\nFirst few rows:")
display(chunk2.head())
```

```python
# Filter the dataframe for dates between 2023-03-30 and 2023-08-03 (inclusive)
chunk3 = df[(df['Date'] >= '2024-06-13') & (df['Date'] <= '2025-02-20')]

# Display the shape and first few rows of the filtered dataframe
print("Shape of filtered dataframe:", chunk3.shape)
print("\nFirst few rows:")
display(chunk3.head())
```

```python
unique_dates = chunk3['Date'].unique()

print("Unique dates:")
print(unique_dates)
```

```python
chunk1.to_csv('chunk1.csv', index=False)
chunk2.to_csv('chunk2.csv', index=False)
chunk3.to_csv('chunk3.csv', index=False)
```

```python
print("Unique countries in chunk1:")
print(len(chunk1['Country'].unique()))

print("\nUnique dates in chunk1:")
print(len(chunk1['Date'].unique()))

print("Unique countries in chunk2:")
print(len(chunk2['Country'].unique()))

print("\nUnique dates in chunk2:")
print(len(chunk2['Date'].unique()))

print("Unique countries in chunk3:")
print(len(chunk3['Country'].unique()))

print("\nUnique dates in chunk3:")
print(len(chunk3['Date'].unique()))
```
