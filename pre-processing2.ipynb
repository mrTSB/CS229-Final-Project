```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv('data132.csv')

unique_dates = len(df['Date'].unique())
date_counts = df['Date'].value_counts()

print("Unique dates:")
print(unique_dates)
print("\nDate counts:")
print(date_counts)
```

```python
# Get value counts of unique countries
country_counts = df['Country'].value_counts()
pd.set_option('display.max_rows', None)

# Display the results
print("Number of unique countries:", len(df['Country'].unique()))
print("\nCountry counts:")
print(country_counts)
```

```python
# Filter the dataframe for the specific date
df_specific_date = df[df['Date'] == '2022-07-22']

# Display the shape and first few rows of the filtered dataframe
print("Shape of filtered dataframe:", df_specific_date.shape)
print("\nFirst few rows:")
display(df_specific_date.head())
```

```python
# Get the unique dates
unique_dates = df['Date'].unique()

print("Unique dates:")
print(unique_dates)
```

```python
# Define the dates to filter
dates_to_limit = ['2022-07-22', '2022-07-21', '2022-07-23', '2022-07-26', 
                  '2022-08-23', '2024-06-06', '2022-09-22', '2023-08-10']

# Filter the dataframe for the specified dates
df_filtered_dates = df[df['Date'].isin(dates_to_limit)]

# Keep only the first 101 rows for the specified dates
df_limited_dates = df_filtered_dates.head(101)

# Get the remaining rows that are not in the specified dates
df_other_rows = df[~df['Date'].isin(dates_to_limit)]

# Concatenate the limited rows with the other rows
df_final = pd.concat([df_limited_dates, df_other_rows])

# Display the shape and first few rows of the final dataframe
print("Shape of final dataframe:", df_final.shape)
print("\nFirst few rows:")
display(df_final.head())
```

```python
unique_dates = len(df_final['Date'].unique())
date_counts = df_final['Date'].value_counts()

print("Unique dates:")
print(unique_dates)
print("\nDate counts:")
print(date_counts)
```

```python
# Get value counts of unique countries
country_counts = df_final['Country'].value_counts()
# Display the results
print("Number of unique countries:", len(df_final['Country'].unique()))
print("\nCountry counts:")
print(country_counts)
```
